```python


from datetime import datetime

batch_id = "Batch_raw_001"
try:
    batch_id = dbutils.widgets.get("batch_id")
except Exception:
    pass

audit_notebook = "audit"
source_layer = "raw"
target_layer = "bronze"

start_time = log_start(
    batch_id,
    audit_notebook,
    source_layer,
    target_layer
)

try:

    # READ DATA
    raw_df = spark.read.format("delta").load(
        "wasbs://bronze@insurancedatahomecredit.blob.core.windows.net/train"
    )

    bronze_df = spark.read.format("delta").load(
        "wasbs://bronze@insurancedatahomecredit.blob.core.windows.net/train"
    )

    # VALIDATION
    raw_count = raw_df.count()
    bronze_count = bronze_df.count()

    if raw_count != bronze_count:
        raise Exception(f"Row mismatch: Raw={raw_count}, Bronze={bronze_count}")

    # SUCCESS LOG
    log_success(
        batch_id,
        audit_notebook,
        source_layer,
        target_layer,
        start_time,
        raw_count,
        bronze_count
    )

except Exception as e:

    log_failure(
        batch_id,
        audit_notebook,
        source_layer,
        target_layer,
        start_time,
        0,
        0,
        str(e)
    )

    raise e
```
